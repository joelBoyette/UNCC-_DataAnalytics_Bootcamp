```python
# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
import numpy as np
import tensorflow as tf
import csv
import urllib
import keras
from keras.preprocessing import image
from keras.applications.vgg19 import (
    VGG19, preprocess_input, decode_predictions)

from PIL import Image
import urllib.request
import io

import pymongo

```

```python
# Load the VGG19 model. https://keras.io/applications/#VGG19
model = VGG19(
    include_top=True,
    weights='imagenet')

# Default Image Size for VGG19 https://keras.io/applications/#vgg19
image_size = (224, 224)
```

```python
dog_meme_df = pd.read_csv('dog_meme.csv',header = None)
dog_meme_df['clean url'] = dog_meme_df[0].str[11:-2]
url_list = list(dog_meme_df['clean url'])
```

```python
# Load the image and resize to the input dimensions that VGG19 was trained with
top_prediction_list = []
image_prediction_dict = {}
image_prediction_list = []
count = 0 

for url in url_list:
     
    with urllib.request.urlopen(url) as url_clean:
        f = io.BytesIO(url_clean.read())
    
    img = image.load_img(f, target_size=image_size)
    
    print(f'{count} of {len(url_list)}')
    # Preprocess image for model prediction
    # This step handles scaling and normalization for VGG19
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Make predictions
    predictions = model.predict(x)
    top_prediction_list = decode_predictions(predictions, top=1)[0]
    
    for prediction in top_prediction_list:
        print(f'{prediction[1]} | {prediction[2]} | {url}')
        image_prediction_dict = {'Prediction': prediction[1],'Prediction_Rank':prediction[2],'Image':url}
        image_prediction_list.append(image_prediction_dict)
    
    count += 1

```

```python
img_df = pd.DataFrame(image_prediction_list)
img_dict = img_df.to_dict('list')
```

```python
dog_unique = img_df['Prediction'].unique()

output_path = os.path.join('dog_list.csv')

with open(output_path,'w',newline='') as csvfile:
    
    csvwriter = csv.writer(csvfile,delimiter=',')
    
    for row in dog_unique:
        csvwriter.writerow([row])
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#creates db (dog_meme_db) and collection (dog_meme_predictions) to hold data
db = client.dog_meme_db
collection = db.dog_meme_predictions

collection.insert_one(img_dict)

```
