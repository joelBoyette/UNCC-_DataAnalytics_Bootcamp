Analysis

1. The max temperature appears to be the highest between 20-25 degrees latitude
2. Humidity is highest at the equator
3. Wind Speed is fairly consistent across latitudes
4. The least amount of clouds occur beteewn 40-60 degrees latitude


```python
import pandas as pd
import random
import openweathermapy.core as ow
from config import api_key
from citipy import citipy
import numpy as np
import requests
import matplotlib.pyplot as plt
from datetime import datetime , date
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
city_name = []
country_code = []

for x in range(2500):
    lat = random.randrange(-90,90)
    lgn = random.randrange(-90,90)

    city_name.append(citipy.nearest_city(lat, lgn).city_name)
    country_code.append(citipy.nearest_city(lat, lgn).country_code)

city_dict = {"City":city_name, "Country":country_code}
city_df = pd.DataFrame(city_dict)
city_df = city_df.drop_duplicates(keep=False)
cities = city_df["City"].tolist()
```

```python
settings = {"units": "Imperial", "appid": api_key}  

weather_data = []
weather_cities = []

print("Beginning Data Retrieval")
print("-------------------------")
for x in range(len(cities)-1):
    try:
        weather_data.append(ow.get_current(cities[x], **settings))
        print(f'Processing record {x+1} | {cities[x]}')
        weather_cities.append(cities[x])
        print(f'{url}&q={cities[x]}&units=Imperial&appid=masked')
    except:
        print(f'Error: Weather not found for {cities[x]}')
```

```python
# Create an "extracts" object for each city
summary = ["main.temp_max","coord.lat","coord.lon","main.humidity","clouds.all", "wind.speed","dt"]
# Create a Pandas DataFrame with the results
data = [response(*summary) for response in weather_data]

column_names = ["Max Temp", "Latitude", "Longitude","Humidity","Cloudiness", "Wind Speed", "Date"]
weather_data = pd.DataFrame(data, index=weather_cities, columns=column_names)

weather_data = weather_data.reset_index()
weather_data = weather_data.rename(columns={"index": "City"})

for index, row in weather_data.iterrows():
    weather_data["Converted Date"] =  pd.to_datetime(row["Date"],unit='s').date()

weather_data = pd.merge(weather_data,city_df[["City","Country"]], on="City",how = "left")
weather_data.to_csv("weather_data.csv")
```

```python
ax = weather_data.plot.scatter("Latitude","Max Temp", s = 5,grid=True,
                        linewidths=.5, edgecolors="black")

plt.title("City Latitude vs Max Temperature (F) 06/22/18")
ax.set_facecolor('#d7d7df')
ax.grid(color="w",  alpha=.7)

plt.savefig(r'images/max_temp_latitude.png')

ax  = weather_data.plot.scatter("Latitude","Humidity", s = 5,grid=True,
                        linewidths=.5, edgecolors="black")

plt.title("City Latitude vs Humidity 06/22/18")
ax.set_facecolor('#d7d7df')
ax.grid(color="w",  alpha=.7)

plt.savefig(r'images/humidity_latitude.png')

ax = weather_data.plot.scatter("Latitude","Cloudiness", s = 5,grid=True,
                        linewidths=.5, edgecolors="black")

plt.title("City Latitude vs Coudiness 06/22/18")
ax.set_facecolor('#d7d7df')
ax.grid(color="w",  alpha=.7)

plt.savefig(r'images/cloudiness_latitude.png')

ax = weather_data.plot.scatter("Latitude","Wind Speed", s = 5,grid=True,
                        linewidths=.5, edgecolors="black")

plt.title("City Latitude vs Wind Speed 06/22/18")
ax.set_facecolor('#d7d7df')
ax.grid(color="w",  alpha=.7)

plt.savefig(r'images/wind_latitude.png')
```
